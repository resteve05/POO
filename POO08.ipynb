```python
import matplotlib.pyplot as plt # esta es la pregunta 5

class Punto(): # definir x y
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def plot(self): #describir la grafica, los ejes y colores
        plt.scatter(self.x, self.y, color='red', marker='o')
        plt.xlabel('X-axis')
        plt.ylabel('Y-axis')
        plt.title('Gráfica del punto')
        plt.grid(True)
        plt.show()
    def plot2(self):
        plt.scatter(self.x, self.y, color='blue', marker='o')
        plt.xlabel('X-axis')
        plt.ylabel('Y-axis')
        plt.title('Gráfica del punto')
        plt.grid(True)
        plt.show()
```

```python
punto1 = Punto(5,2) #definir los puntos
punto2 = Punto(3,4)
```

```python
punto1.plot() #ejecutar los puntos
punto2.plot2() #esta es la pregunta 6, probar el codigo
```

```python
class Computadora(): #definir la clase madre
    def __init__(self, marca, color): #definir variables de la clase madre
        self.marca = marca
        self.color = color

    def encender(self):
        print("Listo para usar la computadora.")

class Mac(Computadora): #clase heredada o subclase
    def __init__(self, marca, color, tamaño): #definir sus variables
        super().__init__(marca, color)
        self.tamaño = tamaño

    def prender(self, prender): #ifs si lo ingresado es igual a esto o no
        if prender == 'Cargador':
            print('Cargador conectado, encienda computadora.')
        else:
            print('No está cargando, cambiar cargador.')

    def apagar(self, apagar):
        if apagar == 'Apagar':
            print('Computadora apagada.')
        else:
            print('Sigue prendida la computadora')

class Android(Computadora): #repeticion de lo primero
    def __init__(self, marca, color, tamaño):
        super().__init__(marca, color)
        self.tamaño = tamaño

    def funcion(self, funcion):
        if funcion == 'Andoid no funciona':
            print('Andoid es chafa')
        else:
            print('Android es muy chafa')

    def correr(self, correr):
        if correr == 'Correr android':
            print('Android es lenta')
        else:
            print('Android es muy lenta')
```

```python
computadora = Computadora("Alguna marca","Algún color") #definir variables de clase madre
```

```python
Computadora.encender(False) #ejecutar clase madre
```
